## Dealing with incorrect registration year date

Because a car can't be first registered after the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate. and any year below 1900 is unrealistic.

```python
autos["registration_year"].value_counts().sort_index(ascending=False)
```

the higest year is 2016 since that is the last year cars were listed based on the crawled column. and the lowest year would be 1900 because that is the century cars where invented. 

```python
autos=autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).sort_index()
```

most of the cars are registered in the last 20 years


## Exploring price by brand.

Exploring the brand column. 

```python
brands = autos["brand"].value_counts()
brands
```

volkswagen has the higest number of car listings and lada the least. i will select the first of 20 popular cars. 

```python
common_brands = brands.head(20).index
common_brands
```

```python
brand_mean_price = {}

for brand in common_brands:
    brandt = autos[autos["brand"] == brand]
    meanprice = brandt["price"].mean()
    brand_mean_price[brand] = meanprice.round()
print(brand_mean_price)
```

the top expensive cars are, sonstige,mini,audi,mercedes_benz,bmw. then volkswagen,skoda,hyundai are in the middle.

## Exploring  mileage 

```python
bmp_series = pd.Series(brand_mean_price)
pd.DataFrame(bmp_series,columns=["mean_price"])
```

```python
brand_mean_mileage = {}
for brand in common_brands:
    brandt = autos[autos["brand"] == brand]
    meanmileage = brandt["odometer_km"].mean()
    brand_mean_mileage[brand] = meanmileage.round()
bmp_series = pd.Series(brand_mean_price).sort_values(ascending=False)
bmm_series = pd.Series(brand_mean_mileage).sort_values(ascending=False)
```
